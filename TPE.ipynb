# Hipotesis

## Hipotesis 1 / Univariada 

#H₀: La proporción de visitas en mayo es igual o menor a 30%.

#H₁: La proporción de visitas en mayo es mayor que 30%.

```python
import math
import numpy as np
from scipy import stats

total = raw_dataset_mod["Month"].count()
conteo_por_mes = raw_dataset_mod["Month"].value_counts()
Mayo_total = conteo_por_mes["May"]
p0=0.30 
p_obs=Mayo_total/total 

res_binom = stats.binomtest(Mayo_total, total, p=p0, alternative='greater')
p_value_binom = res_binom.pvalue

se = np.sqrt(p0 * (1 - p0) / total)
z = (p_obs - p0) / se 
p_value_z = stats.norm.sf(z)

alpha = 0.05
print()

print(f"Total: {total}")
print(f"Visitas en May: {Mayo_total}")
print(f"Proporción observada: {p_obs:.4f}")
print(f"Proporción H0 (p0): {p0:.2f}")
print(f"Estadístico z: {z:.4f}")
if p_value_binom < alpha:
    print(f"Binomial: Rechazamos H0 al nivel {alpha:.2f} (p={p_value_binom:.4f})")
else:
    print(f"Binomial: No rechazamos H0 al nivel {alpha:.2f} (p={p_value_binom:.4f})")

if p_value_z < alpha:
    print(f"Z-test: Rechazamos H0 al nivel {alpha:.2f} (p={p_value_z:.4f})")
else:
    print(f"Z-test: No rechazamos H0 al nivel {alpha:.2f} (p={p_value_z:.4f})")
```

## Hipotesis 2 / Bivariada 

H₀: El tipo de cliente es independiente de la compra.

H₁: Existe relación entre tipo de cliente y la compra.

```python
import math
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency

tabla = pd.crosstab(raw_dataset_mod["VisitorType"], raw_dataset_mod["Revenue"])
chi2, p_value, df, esperado = chi2_contingency(tabla)

print("Chi2 calculado:", chi2)
print("p-value:", p_value)
print("Grados de libertad:", df)
print("Frecuencias observadas:\n", tabla.values)
print("Frecuencias esperadas:\n", esperado)

alpha = 0.05
if p_value < alpha:
    print("\nConclusión: Se RECHAZA H0 → Existe relación entre VisitorType y Revenue.")
else:
    print("\nConclusión: NO se rechaza H0 → Son independientes.")
```

## Hipotesis 3 / Univariada 

H₀: El promedio de BR es igual o menor al umbral de 2%.

H₁: El promedio de BR es mayor que el umbral de 2%.

```python
from os import path, listdir
import numpy as np
from scipy import stats

bounce = raw_dataset_mod["BounceRates"].dropna().astype(float)
n = len(bounce)                
x_bar = bounce.mean()          
s = bounce.std(ddof=1)          
mu0 = 2 

se = s / np.sqrt(n)                    
t_stat = (x_bar - mu0) / se               

p_one_sided = stats.t.sf(t_stat, df=n-1)
t_stat_scipy, p_two_sided = stats.ttest_1samp(bounce, mu0, nan_policy="omit")

p_one_sided_from_ttest = p_two_sided / 2 if t_stat_scipy > 0 else 1 - p_two_sided / 2
alpha = 0.05

print(f"Tamaño de la muestra (n): {n}")
print(f"Media muestral (x̄): {x_bar:.4f}")
print(f"Desviación est. muestral (s): {s:.4f}")
print(f"Estadístico t (manual): {t_stat:.4f}")
print(f"p-valor unilateral (usando stats.t.sf): {p_one_sided:.6f}")

if p_one_sided < alpha:
    print(f"Conclusión: RECHAZAMOS H0 al nivel {alpha:.2f} (p = {p_one_sided:.6f}) → El promedio de BR es significativamente mayor que {mu0:.3f}.")
else:
    print(f"Conclusión: NO RECHAZAMOS H0 al nivel {alpha:.2f} (p = {p_one_sided:.6f}) → No hay evidencia suficiente de que el promedio de BR sea mayor que {mu0:.3f}.")
```

## Hipotesis 6 / Bivariada

H₀: no hay diferencias en la distribución (mediana) de PageValues entre compradores y no compradores. 

H₁: la mediana de PageValues es mayor en las sesiones donde finalmente se realiza una compra.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene, mannwhitneyu

compran = raw_dataset_mod[raw_dataset_mod["Revenue"] == True]["PageValues"]
no_compran = raw_dataset_mod[raw_dataset_mod["Revenue"] == False]["PageValues"]

# Test de normalidad (Shapiro-Wilk) -> Shapiro se aplica sobre muestras cuando N es muy grande
sample_compran = compran.sample(500, random_state=42)
sample_no_compran = no_compran.sample(500, random_state=42)
print()
print("Test de normalidad Shapiro-Wilk")
print("Compran:", shapiro(sample_compran))
print("No compran:",shapiro(sample_no_compran))
print()

# Test de homogeneidad de varianzas (Levene)
print("Test de homogeneidad de varianzas (Levene)")
print (levene(compran, no_compran))
print()

# Test Mann-Whitney U (bilateral)
mw = mannwhitneyu(compran, no_compran, alternative="two-sided")
print()
print("Test Mann-Whitney U (bilateral)")
print(mw)

alpha = 0.05

print()
print("Conclusión de la hipótesis")
if mw.pvalue < alpha:
    print(f"Se rechaza H0 (p-value = {mw.pvalue:.4f}).")
    print("Las distribuciones de PageValues difieren significativamente entre ambos grupos.")
else:
    print(f"No se puede rechazar H0 (p-value = {mw.pvalue:.4f}).")
    print("No hay evidencia estadística de diferencias en PageValues entre los grupos.")
```
